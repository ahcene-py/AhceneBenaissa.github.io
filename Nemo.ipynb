```python
# Importer les bibliothèques nécessaires
import nemo.collections.asr as nemo_asr
```

```python
# Charger le modèle de reconnaissance vocale en français
print(" Téléchargement du modèle de reconnaissance vocale...")
asr_model = nemo_asr.models.ASRModel.from_pretrained(model_name="stt_fr_conformer_ctc_large")
```

```python
# Définir la fonction pour convertir l'audio en texte
def speech_to_text(audio_file):
    print("Conversion de l'audio en texte...")
    transcription = asr_model.transcribe([audio_file])[0]
    print(" Transcription :", transcription)
    return transcription
```

```python
# Récupérer le nom du fichier audio
audio_file = list(uploaded.keys())[0]

# Lancer la transcription
transcribed_text = speech_to_text(audio_file)
```
